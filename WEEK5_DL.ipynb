```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
def create_base_model():
    model = models.Sequential([
        layers.Dense(256, activation='relu', input_shape=(784,)),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
model = create_base_model()

history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2
)

test_loss, test_acc = model.evaluate(x_test, y_test)
print("Base Model Test Accuracy:", test_acc)
```

```python
model_l2 = models.Sequential([
    layers.Dense(256,activation='relu',kernel_regularizer=regularizers.l2(0.001),input_shape=(784,)),
    layers.Dense(128,activation='relu',kernel_regularizer=regularizers.l2(0.001)),
    layers.Dense(10,activation='softmax')
])

model_l2.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

model_l2.fit(x_train,y_train,epochs=10,batch_size=128,validation_split=0.2)

loss,acc = model_l2.evaluate(x_test,y_test)

print("L2 Regularization Accuracy:",acc)
```

```python
x_train_img = x_train.reshape(-1,28,28,1)
x_test_img = x_test.reshape(-1,28,28,1)

datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1
)

model_aug = models.Sequential([
    layers.Flatten(input_shape=(28,28,1)),
    layers.Dense(256,activation='relu'),
    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])

model_aug.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

model_aug.fit(datagen.flow(x_train_img,y_train,batch_size=128),
              epochs=10,
              validation_data=(x_test_img,y_test))

loss,acc = model_aug.evaluate(x_test_img,y_test)

print("Dataset Augmentation Accuracy:",acc)
```

```python
def create_shared_model():
    inputs = tf.keras.layers.Input(shape=(784,))

    x = tf.keras.layers.Dense(128, activation='relu')(inputs)

    shared_layer = tf.keras.layers.Dense(128, activation='relu')

    x = shared_layer(x)
    x = shared_layer(x)

    outputs = tf.keras.layers.Dense(10, activation='softmax')(x)

    model = tf.keras.Model(inputs, outputs)

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
model_shared = create_shared_model()

history_shared = model_shared.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2
)

test_loss, test_acc = model_shared.evaluate(x_test, y_test)

print("Shared Parameter Model Test Accuracy:", test_acc)
```

```python
model_noise = models.Sequential([
    layers.GaussianNoise(0.1,input_shape=(784,)),
    layers.Dense(256,activation='relu'),
    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])

model_noise.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

model_noise.fit(x_train,y_train,epochs=10,batch_size=128,validation_split=0.2)

loss,acc = model_noise.evaluate(x_test,y_test)

print("Noise Injection Accuracy:",acc)
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

model_es = models.Sequential([
    layers.Dense(256,activation='relu',input_shape=(784,)),
    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])

model_es.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

model_es.fit(x_train,y_train,epochs=30,batch_size=128,
             validation_split=0.2,
             callbacks=[early_stop])

loss,acc = model_es.evaluate(x_test,y_test)

print("Early Stopping Accuracy:",acc)
```

```python
model_dropout = models.Sequential([
    layers.Dense(256,activation='relu',input_shape=(784,)),
    layers.Dropout(0.5),
    layers.Dense(128,activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10,activation='softmax')
])

model_dropout.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

model_dropout.fit(x_train,y_train,epochs=10,batch_size=128,validation_split=0.2)

loss,acc = model_dropout.evaluate(x_test,y_test)

print("Dropout Accuracy:",acc)
```

```python
def ensemble_predict(models, x):
    predictions = [model.predict(x) for model in models]
    avg_prediction = np.mean(predictions, axis=0)
    return np.argmax(avg_prediction, axis=1)

model1 = create_dropout_model()
model2 = create_l2_model()

model1.fit(x_train, y_train, epochs=5, batch_size=128)
model2.fit(x_train, y_train, epochs=5, batch_size=128)

ensemble_result = ensemble_predict([model1, model2], x_test)
```
