```python
import numpy as np
np.random.seed(0)

X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])

def sigmoid(x):
    return 1/(1+np.exp(-x))

def dsigmoid(x):
    return x*(1-x)

class MLP:
    def __init__(self):
        self.W1 = np.random.randn(2,2)
        self.b1 = np.zeros((1,2))
        self.W2 = np.random.randn(2,1)
        self.b2 = np.zeros((1,1))

    def forward(self,X):
        self.z1 = np.dot(X,self.W1)+self.b1
        self.a1 = sigmoid(self.z1)
        self.z2 = np.dot(self.a1,self.W2)+self.b2
        self.a2 = sigmoid(self.z2)
        return self.a2

    def backward(self,X,Y):
        m = X.shape[0]
        dz2 = self.a2 - Y
        dW2 = np.dot(self.a1.T,dz2)/m
        db2 = np.sum(dz2,axis=0,keepdims=True)/m
        dz1 = np.dot(dz2,self.W2.T)*dsigmoid(self.a1)
        dW1 = np.dot(X.T,dz1)/m
        db1 = np.sum(dz1,axis=0,keepdims=True)/m
        return dW1,db1,dW2,db2

def train_bgd(model,epochs=10000,lr=0.1):
    for i in range(epochs):
        model.forward(X)
        dW1,db1,dW2,db2 = model.backward(X,Y)
        model.W1 -= lr*dW1
        model.b1 -= lr*db1
        model.W2 -= lr*dW2
        model.b2 -= lr*db2

model = MLP()
train_bgd(model)

print("BGD Output:")
print(np.round(model.forward(X)))
```

```python
def train_sgd(model,epochs=10000,lr=0.1):
    for i in range(epochs):
        for j in range(len(X)):
            x = X[j:j+1]
            y = Y[j:j+1]
            model.forward(x)
            dW1,db1,dW2,db2 = model.backward(x,y)
            model.W1 -= lr*dW1
            model.b1 -= lr*db1
            model.W2 -= lr*dW2
            model.b2 -= lr*db2

model = MLP()
train_sgd(model)

print("SGD Output:")
print(np.round(model.forward(X)))
```

```python
def train_minibatch(model,epochs=10000,lr=0.1,batch=2):
    for i in range(epochs):
        for j in range(0,len(X),batch):
            x = X[j:j+batch]
            y = Y[j:j+batch]
            model.forward(x)
            dW1,db1,dW2,db2 = model.backward(x,y)
            model.W1 -= lr*dW1
            model.b1 -= lr*db1
            model.W2 -= lr*dW2
            model.b2 -= lr*db2

model = MLP()
train_minibatch(model)

print("MiniBatch Output:")
print(np.round(model.forward(X)))
```

```python
def train_momentum(model,epochs=10000,lr=0.1,beta=0.9):
    vW1=np.zeros_like(model.W1)
    vb1=np.zeros_like(model.b1)
    vW2=np.zeros_like(model.W2)
    vb2=np.zeros_like(model.b2)

    for i in range(epochs):
        model.forward(X)
        dW1,db1,dW2,db2 = model.backward(X,Y)

        vW1 = beta*vW1 + (1-beta)*dW1
        vb1 = beta*vb1 + (1-beta)*db1
        vW2 = beta*vW2 + (1-beta)*dW2
        vb2 = beta*vb2 + (1-beta)*db2

        model.W1 -= lr*vW1
        model.b1 -= lr*vb1
        model.W2 -= lr*vW2
        model.b2 -= lr*vb2

model = MLP()
train_momentum(model)

print("Momentum Output:")
print(np.round(model.forward(X)))
```

```python
def train_nesterov(model,epochs=10000,lr=0.1,beta=0.9):
    vW1=np.zeros_like(model.W1)
    vb1=np.zeros_like(model.b1)
    vW2=np.zeros_like(model.W2)
    vb2=np.zeros_like(model.b2)

    for i in range(epochs):
        W1_temp = model.W1 - beta*vW1
        W2_temp = model.W2 - beta*vW2

        model.forward(X)
        dW1,db1,dW2,db2 = model.backward(X,Y)

        vW1 = beta*vW1 + lr*dW1
        vb1 = beta*vb1 + lr*db1
        vW2 = beta*vW2 + lr*dW2
        vb2 = beta*vb2 + lr*db2

        model.W1 -= vW1
        model.b1 -= vb1
        model.W2 -= vW2
        model.b2 -= vb2

model = MLP()
train_nesterov(model)

print("Nesterov Output:")
print(np.round(model.forward(X)))
```

```python
def train_adagrad(model,epochs=10000,lr=0.5,eps=1e-8):
    sW1=np.zeros_like(model.W1)
    sb1=np.zeros_like(model.b1)
    sW2=np.zeros_like(model.W2)
    sb2=np.zeros_like(model.b2)

    for i in range(epochs):
        model.forward(X)
        dW1,db1,dW2,db2 = model.backward(X,Y)

        sW1 += dW1**2
        sb1 += db1**2
        sW2 += dW2**2
        sb2 += db2**2

        model.W1 -= lr*dW1/(np.sqrt(sW1)+eps)
        model.b1 -= lr*db1/(np.sqrt(sb1)+eps)
        model.W2 -= lr*dW2/(np.sqrt(sW2)+eps)
        model.b2 -= lr*db2/(np.sqrt(sb2)+eps)

model = MLP()
train_adagrad(model)

print("Adagrad Output:")
print(np.round(model.forward(X)))
```

```python
def train_adadelta(model,epochs=10000,rho=0.95,eps=1e-6):
    EgW1=np.zeros_like(model.W1)
    EgW2=np.zeros_like(model.W2)
    EdW1=np.zeros_like(model.W1)
    EdW2=np.zeros_like(model.W2)

    for i in range(epochs):
        model.forward(X)
        dW1,db1,dW2,db2 = model.backward(X,Y)

        EgW1 = rho*EgW1 + (1-rho)*dW1**2
        EgW2 = rho*EgW2 + (1-rho)*dW2**2

        updateW1 = -np.sqrt(EdW1+eps)/np.sqrt(EgW1+eps)*dW1
        updateW2 = -np.sqrt(EdW2+eps)/np.sqrt(EgW2+eps)*dW2

        model.W1 += updateW1
        model.W2 += updateW2

model = MLP()
train_adadelta(model)

print("Adadelta Output:")
print(np.round(model.forward(X)))
```

```python
def train_adam(model,epochs=10000,lr=0.01,b1=0.9,b2=0.999,eps=1e-8):
    mW1=np.zeros_like(model.W1)
    vW1=np.zeros_like(model.W1)
    mW2=np.zeros_like(model.W2)
    vW2=np.zeros_like(model.W2)

    for t in range(1,epochs+1):
        model.forward(X)
        dW1,db1,dW2,db2 = model.backward(X,Y)

        mW1 = b1*mW1 + (1-b1)*dW1
        vW1 = b2*vW1 + (1-b2)*(dW1**2)

        mW1_hat = mW1/(1-b1**t)
        vW1_hat = vW1/(1-b2**t)

        model.W1 -= lr*mW1_hat/(np.sqrt(vW1_hat)+eps)

model = MLP()
train_adam(model)

print("Adam Output:")
print(np.round(model.forward(X)))
```

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

np.random.seed(0)

data = load_breast_cancer()
X = data.data
Y = data.target.reshape(-1,1)

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)

def sigmoid(x):
    return 1/(1+np.exp(-x))

def dsigmoid(x):
    return x*(1-x)

class MLP:
    def __init__(self,input_size,hidden,output):
        self.W1 = np.random.randn(input_size,hidden)*0.01
        self.b1 = np.zeros((1,hidden))
        self.W2 = np.random.randn(hidden,output)*0.01
        self.b2 = np.zeros((1,output))

    def forward(self,X):
        self.z1 = np.dot(X,self.W1)+self.b1
        self.a1 = sigmoid(self.z1)
        self.z2 = np.dot(self.a1,self.W2)+self.b2
        self.a2 = sigmoid(self.z2)
        return self.a2

    def backward(self,X,Y):
        m = X.shape[0]
        dz2 = self.a2 - Y
        dW2 = np.dot(self.a1.T,dz2)/m
        db2 = np.sum(dz2,axis=0,keepdims=True)/m
        dz1 = np.dot(dz2,self.W2.T)*dsigmoid(self.a1)
        dW1 = np.dot(X.T,dz1)/m
        db1 = np.sum(dz1,axis=0,keepdims=True)/m
        return dW1,db1,dW2,db2

def accuracy(model):
    preds = np.round(model.forward(X_test))
    return np.mean(preds==Y_test)*100
```

```python
model = MLP(30,16,1)
for i in range(2000):
    model.forward(X_train)
    dW1,db1,dW2,db2 = model.backward(X_train,Y_train)
    model.W1 -= 0.1*dW1
    model.W2 -= 0.1*dW2
print("BGD:",accuracy(model))
```
