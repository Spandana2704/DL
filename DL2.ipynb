```python
def AND(x1,x2):
    w1=1
    w2=1
    b=-1.5
    activation=w1*x1+w1*x2+b
    if activation>=0:
        return 1
    else:
        return 0

def OR(x1,x2):
    w1=1
    w2=1
    b=-0.5
    activation=w1*x1+w1*x2+b
    if activation>=0:
        return 1
    else:
        return 0

def NAND(x1,x2):
    w1=-1
    w2=-1
    b=1.5
    activation=w1*x1+w1*x2+b
    if activation>=0:
        return 1
    else:
        return 0

def XOR(x1,x2):
    first_layer_op=NAND(x1,x2)
    sec_layer_op=OR(x1,x2)
    op=AND(first_layer_op,sec_layer_op)
    return op

x=[(0,0),(0,1),(1,0),(1,1)]
for x1,x2 in x:
    print(XOR(x1,x2),end=" ")
```

```python
def NXOR(x1,x2):
    first_layer_op=NAND(x1,x2)
    sec_layer_op=OR(x1,x2)
    op=NAND(first_layer_op,sec_layer_op)
    return op

x=[(0,0),(0,1),(1,0),(1,1)]
for x1,x2 in x:
    print(NXOR(x1,x2),end=" ")
```
